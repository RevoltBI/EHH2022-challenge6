```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open("grads.data", "rb") as f:
    data = pickle.load(f)
```

```python
print(data[0].shape)

res = np.zeros((40, 598))

for x in data:
    for y in x:
        res += np.abs(y)
```

```python
sms = res.sum(axis=0).reshape((2, -1))
```

```python
for i in range(6):
    plt.figure(figsize=(30, 1))
    plt.imshow(sms[:, (i * 50):((i+1)*50)])
    plt.show()
```
